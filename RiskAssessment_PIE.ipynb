```python
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np

vehicle_xml_path = "/content/PIE/vehicle_annotations_unzipped/annotations_vehicle/set03/video_0006_obd.xml"

tree = ET.parse(vehicle_xml_path)
root = tree.getroot()

vehicle_rows = []

for frame in root.findall("frame"):
    a = frame.attrib

    frame_id = int(a["id"])

    def safe_float(key, default=0.0):
        try:
            return float(a.get(key, default))
        except:
            return default

    vehicle_rows.append({
        "frame": frame_id,
        "GPS_speed": safe_float("GPS_speed"),
        "OBD_speed": safe_float("OBD_speed"),
        "gyroZ": safe_float("gyroZ"),
        "heading_angle": safe_float("heading_angle"),
        "yaw": safe_float("yaw"),
        "accX": safe_float("accX"),
        "accY": safe_float("accY"),
        "accZ": safe_float("accZ")
    })

df_vehicle = pd.DataFrame(vehicle_rows)

# Calculate heading change as a steering/turning proxy
df_vehicle["heading_change"] = df_vehicle["heading_angle"].diff().fillna(0)

# Fix angle wraparound, e.g. 359 to 0
df_vehicle["heading_change"] = ((df_vehicle["heading_change"] + 180) % 360) - 180

df_vehicle.head()
```

```python
vehicle_data = {}

for _, row in df_vehicle.iterrows():
    vehicle_data[int(row["frame"])] = {
        "GPS_speed": float(row["GPS_speed"]),
        "OBD_speed": float(row["OBD_speed"]),
        "gyroZ": float(row["gyroZ"]),
        "heading_angle": float(row["heading_angle"]),
        "heading_change": float(row["heading_change"]),
        "yaw": float(row["yaw"])
    }

def get_vehicle_state(frame_id):
    if frame_id in vehicle_data:
        return vehicle_data[frame_id]

    # fallback: nearest frame
    nearest_frame = min(vehicle_data.keys(), key=lambda x: abs(x - frame_id))
    return vehicle_data[nearest_frame]
```

```python
import cv2
import os
import numpy as np
from ultralytics import YOLO, SAM
from google.colab.patches import cv2_imshow

# =============================
# PATHS
# =============================
video_path = "/content/PIE/annotations/PIE_clips/set03/video_0006.mp4"

output_folder = "/content/PIE_video_0006_risk_no_tracking_ego_motion/"
os.makedirs(output_folder, exist_ok=True)

# =============================
# SETTINGS
# =============================
start_frame = 6870
num_frames_to_process = 622

conf_threshold = 0.50
iou_threshold = 0.50

frame_stride = 30
show_first_n = 25

# =============================
# MODELS
# =============================
yolo_model = YOLO("yolov8m.pt")
sam_model = SAM("sam2.1_s.pt")


# =============================
# RISK FUNCTION WITHOUT TRACKING
# =============================
def calculate_risk_no_tracking(
    x1, y1, x2, y2,
    frame_w, frame_h,
    vehicle_state
):
    box_w = max(x2 - x1, 1)
    box_h = max(y2 - y1, 1)
    box_area = box_w * box_h
    image_area = frame_w * frame_h

    cx = (x1 + x2) / 2
    cy = (y1 + y2) / 2

    # -----------------------------
    # 1. Pedestrian image-space features
    # -----------------------------
    area_ratio = box_area / image_area

    # Stronger proximity control:
    # far/small pedestrians get low proximity.
    proximity = min(area_ratio / 0.045, 1.0)

    vertical_closeness = cy / frame_h

    center_distance = abs(cx - frame_w / 2) / (frame_w / 2)
    center_factor = 1.0 - min(center_distance, 1.0)

    # Approximate ego path zone
    ego_x1 = frame_w * 0.30
    ego_x2 = frame_w * 0.70
    ego_y1 = frame_h * 0.55

    in_ego_path = (ego_x1 <= cx <= ego_x2) and (cy >= ego_y1)

    # Pedestrian side: left negative, right positive
    pedestrian_side = (cx - frame_w / 2) / (frame_w / 2)

    # -----------------------------
    # 2. Ego vehicle speed and turning
    # -----------------------------
    gps_speed = vehicle_state["GPS_speed"]
    obd_speed = vehicle_state["OBD_speed"]
    gyroZ = vehicle_state["gyroZ"]
    heading_change = vehicle_state["heading_change"]

    # Use GPS if OBD is zero or unreliable
    ego_speed = gps_speed if gps_speed > 0.1 else obd_speed
    ego_speed_factor = min(ego_speed / 50.0, 1.0)

    # Turning proxy, not true steering wheel angle
    turn_proxy = heading_change + 0.2 * gyroZ
    turn_intensity = min(abs(turn_proxy) / 4.0, 1.0)

    if turn_proxy > 0:
        turning_side = 1
    elif turn_proxy < 0:
        turning_side = -1
    else:
        turning_side = 0

    ped_side_sign = 1 if pedestrian_side > 0 else -1

    # Higher only if ego vehicle turns toward pedestrian side
    turning_toward_ped = 1.0 if turning_side != 0 and ped_side_sign == turning_side else 0.0
    steering_factor = turn_intensity * turning_toward_ped

    # -----------------------------
    # 3. Distance gates
    # -----------------------------
    very_far = proximity < 0.12 and vertical_closeness < 0.55
    far = proximity < 0.22 and vertical_closeness < 0.65
    close = proximity > 0.35 or vertical_closeness > 0.72

    # -----------------------------
    # 4. Risk score
    # -----------------------------
    raw_score = (
        0.42 * proximity +
        0.20 * vertical_closeness +
        0.10 * center_factor +
        0.14 * ego_speed_factor +
        0.14 * steering_factor
    )

    # Ego path relevance
    if in_ego_path:
        raw_score += 0.15

    # Ego speed matters more only if pedestrian is close or in path
    if ego_speed_factor > 0.35 and (close or in_ego_path):
        raw_score += 0.08

    # Steering matters only if not very far
    if steering_factor > 0.2 and not very_far:
        raw_score += 0.08 * steering_factor

    # -----------------------------
    # 5. Risk caps to fix far pedestrian problem
    # -----------------------------
    if very_far:
        raw_score = min(raw_score, 0.34)   # LOW max

    elif far and not in_ego_path:
        raw_score = min(raw_score, 0.45)   # LOW/MEDIUM max

    elif far and in_ego_path:
        raw_score = min(raw_score, 0.58)   # MEDIUM max

    # HIGH only for close or ego-path relevant pedestrians
    if not close and not in_ego_path:
        raw_score = min(raw_score, 0.59)

    risk_score = min(max(raw_score, 0.0), 1.0)

    if risk_score >= 0.80:
        risk_label = "CRITICAL"
        color = (0, 0, 255)
    elif risk_score >= 0.60:
        risk_label = "HIGH"
        color = (0, 165, 255)
    elif risk_score >= 0.40:
        risk_label = "MEDIUM"
        color = (0, 255, 255)
    else:
        risk_label = "LOW"
        color = (0, 255, 0)

    debug = {
        "proximity": proximity,
        "vertical": vertical_closeness,
        "center": center_factor,
        "ego_speed": ego_speed,
        "ego_speed_factor": ego_speed_factor,
        "turn_proxy": turn_proxy,
        "steering_factor": steering_factor,
        "in_ego_path": in_ego_path,
        "very_far": very_far,
        "far": far,
        "close": close
    }

    return risk_score, risk_label, color, debug


# =============================
# PROCESS VIDEO
# =============================
cap = cv2.VideoCapture(video_path)
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

local_frame_idx = 0
shown_count = 0
saved_count = 0

print("Running risk assessment WITHOUT tracking...")

while local_frame_idx < num_frames_to_process:
    ret, frame = cap.read()

    if not ret:
        print("Could not read frame:", local_frame_idx)
        break

    original_frame = start_frame + local_frame_idx

    if local_frame_idx % frame_stride == 0:
        frame_h, frame_w = frame.shape[:2]
        vehicle_state = get_vehicle_state(original_frame)

        temp_path = "/content/temp_risk_no_tracking.jpg"
        cv2.imwrite(temp_path, frame)

        # -----------------------------
        # YOLO detection only, no tracking
        # -----------------------------
        yolo_results = yolo_model(
            temp_path,
            classes=[0],
            conf=conf_threshold,
            iou=iou_threshold,
            imgsz=1280,
            verbose=False
        )

        boxes = yolo_results[0].boxes.xyxy.cpu().numpy() if yolo_results[0].boxes is not None else []
        confs = yolo_results[0].boxes.conf.cpu().numpy() if yolo_results[0].boxes is not None else []

        output = frame.copy()

        # Draw approximate ego path
        ego_x1 = int(frame_w * 0.30)
        ego_x2 = int(frame_w * 0.70)
        ego_y1 = int(frame_h * 0.55)

        cv2.rectangle(output, (ego_x1, ego_y1), (ego_x2, frame_h), (255, 255, 255), 2)
        cv2.putText(
            output,
            "Approx. ego path",
            (ego_x1, ego_y1 - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (255, 255, 255),
            2
        )

        # Vehicle info
        turn_proxy = vehicle_state["heading_change"] + 0.2 * vehicle_state["gyroZ"]

        ego_text = (
            f"GPS: {vehicle_state['GPS_speed']:.1f} | "
            f"OBD: {vehicle_state['OBD_speed']:.1f} | "
            f"turn proxy: {turn_proxy:.2f}"
        )

        cv2.putText(
            output,
            ego_text,
            (20, 35),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (255, 255, 255),
            3
        )

        # -----------------------------
        # SAM2 segmentation
        # -----------------------------
        masks = []

        if len(boxes) > 0:
            sam_results = sam_model(
                temp_path,
                bboxes=np.array(boxes),
                verbose=False
            )

            if sam_results[0].masks is not None:
                masks = sam_results[0].masks.data.cpu().numpy()

        # -----------------------------
        # Risk per detected pedestrian
        # -----------------------------
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = box.astype(int)
            conf = float(confs[i]) if i < len(confs) else 0.0

            risk_score, risk_label, color, debug = calculate_risk_no_tracking(
                x1, y1, x2, y2,
                frame_w, frame_h,
                vehicle_state
            )

            # Mask overlay
            if i < len(masks):
                mask_bool = masks[i].astype(bool)
                colored = output.copy()
                colored[mask_bool] = color
                output = cv2.addWeighted(output, 0.75, colored, 0.25, 0)

            # Box
            cv2.rectangle(output, (x1, y1), (x2, y2), color, 3)

            # Centroid
            cx = int((x1 + x2) / 2)
            cy = int((y1 + y2) / 2)
            cv2.circle(output, (cx, cy), 5, color, -1)

            label = f"{risk_label} {risk_score:.2f} | conf {conf:.2f}"

            cv2.putText(
                output,
                label,
                (x1, max(y1 - 10, 25)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.65,
                (0, 0, 0),
                4
            )

            cv2.putText(
                output,
                label,
                (x1, max(y1 - 10, 25)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.65,
                color,
                2
            )

            debug_text = (
                f"P:{debug['proximity']:.2f} "
                f"E:{debug['ego_speed_factor']:.2f} "
                f"T:{debug['steering_factor']:.2f}"
            )

            cv2.putText(
                output,
                debug_text,
                (x1, min(y2 + 20, frame_h - 10)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.55,
                color,
                2
            )

        save_path = f"{output_folder}risk_no_tracking_frame_{original_frame}.jpg"
        cv2.imwrite(save_path, output)
        saved_count += 1

        print(
            f"Frame {original_frame}: detections={len(boxes)}, "
            f"GPS={vehicle_state['GPS_speed']:.1f}, "
            f"OBD={vehicle_state['OBD_speed']:.1f}, "
            f"turn={turn_proxy:.2f}"
        )

        if shown_count < show_first_n:
            cv2_imshow(output)
            shown_count += 1

    local_frame_idx += 1

cap.release()

print("\nDone.")
print("Shown images:", shown_count)
print("Saved images:", saved_count)
print("Output folder:", output_folder)
```

```python
import cv2
import os
import numpy as np
from ultralytics import YOLO, SAM
from google.colab.patches import cv2_imshow

# =============================
# PATHS
# =============================
video_path = "/content/PIE/annotations/PIE_clips/set03/video_0006.mp4"

output_folder = "/content/PIE_video_0006_dynamic_risk_no_tracking/"
os.makedirs(output_folder, exist_ok=True)

# =============================
# SETTINGS
# =============================
start_frame = 6870
num_frames_to_process = 622

conf_threshold = 0.50
iou_threshold = 0.50

frame_stride = 30
show_first_n = 25

# If dynamic path shifts wrong side, change this to -1
TURN_SIGN = 1

# =============================
# MODELS
# =============================
yolo_model = YOLO("yolov8m.pt")
sam_model = SAM("sam2.1_s.pt")


# =============================
# RISK FUNCTION
# =============================
def calculate_risk_no_tracking_dynamic_path(
    x1, y1, x2, y2,
    frame_w, frame_h,
    vehicle_state
):
    box_w = max(x2 - x1, 1)
    box_h = max(y2 - y1, 1)
    box_area = box_w * box_h
    image_area = frame_w * frame_h

    cx = (x1 + x2) / 2
    cy = (y1 + y2) / 2

    area_ratio = box_area / image_area
    proximity = min(area_ratio / 0.035, 1.0)

    vertical_closeness = cy / frame_h

    center_distance = abs(cx - frame_w / 2) / (frame_w / 2)
    center_factor = 1.0 - min(center_distance, 1.0)

    pedestrian_side = (cx - frame_w / 2) / (frame_w / 2)

    gps_speed = vehicle_state["GPS_speed"]
    obd_speed = vehicle_state["OBD_speed"]
    gyroZ = vehicle_state["gyroZ"]
    heading_change = vehicle_state["heading_change"]

    ego_speed = gps_speed if gps_speed > 0.1 else obd_speed
    ego_speed_factor = min(ego_speed / 50.0, 1.0)

    # Turning proxy, not direct steering wheel angle
    turn_proxy = heading_change + 0.2 * gyroZ
    turn_norm = np.clip((turn_proxy / 4.0) * TURN_SIGN, -1.0, 1.0)

    # Dynamic ego-path region
    base_center = frame_w * 0.50
    path_shift = turn_norm * frame_w * 0.18
    dynamic_center = base_center + path_shift

    base_width = frame_w * 0.40
    turn_extra_width = abs(turn_norm) * frame_w * 0.18
    path_width = base_width + turn_extra_width

    ego_x1 = max(0, dynamic_center - path_width / 2)
    ego_x2 = min(frame_w, dynamic_center + path_width / 2)
    ego_y1 = frame_h * 0.50

    in_dynamic_ego_path = (ego_x1 <= cx <= ego_x2) and (cy >= ego_y1)

    if turn_norm > 0.10:
        turning_side = 1
    elif turn_norm < -0.10:
        turning_side = -1
    else:
        turning_side = 0

    ped_side_sign = 1 if pedestrian_side > 0 else -1
    on_turning_side = turning_side != 0 and ped_side_sign == turning_side

    very_far = proximity < 0.10 and vertical_closeness < 0.55
    far = proximity < 0.20 and vertical_closeness < 0.65
    close = proximity > 0.30 or vertical_closeness > 0.70
    very_close = proximity > 0.42 or vertical_closeness > 0.78

    raw_score = (
        0.45 * proximity +
        0.22 * vertical_closeness +
        0.08 * center_factor +
        0.15 * ego_speed_factor +
        0.10 * abs(turn_norm)
    )

    if in_dynamic_ego_path:
        raw_score += 0.18

    if on_turning_side and not very_far:
        raw_score += 0.15

    if ego_speed_factor > 0.25 and (close or in_dynamic_ego_path or on_turning_side):
        raw_score += 0.10

    # Strong escalation rules
    if close and in_dynamic_ego_path:
        raw_score = max(raw_score, 0.65)

    if very_close and (in_dynamic_ego_path or on_turning_side):
        raw_score = max(raw_score, 0.70)

    if close and on_turning_side and abs(turn_norm) > 0.15:
        raw_score = max(raw_score, 0.68)

    if very_close and on_turning_side and ego_speed_factor > 0.25:
        raw_score = max(raw_score, 0.80)

    # Caps for far pedestrians
    if very_far:
        raw_score = min(raw_score, 0.34)

    elif far and not in_dynamic_ego_path:
        raw_score = min(raw_score, 0.45)

    elif far and in_dynamic_ego_path:
        raw_score = min(raw_score, 0.58)

    risk_score = min(max(raw_score, 0.0), 1.0)

    if risk_score >= 0.80:
        risk_label = "CRITICAL"
        color = (0, 0, 255)
    elif risk_score >= 0.60:
        risk_label = "HIGH"
        color = (0, 165, 255)
    elif risk_score >= 0.40:
        risk_label = "MEDIUM"
        color = (0, 255, 255)
    else:
        risk_label = "LOW"
        color = (0, 255, 0)

    debug = {
        "proximity": proximity,
        "vertical": vertical_closeness,
        "ego_speed": ego_speed,
        "ego_speed_factor": ego_speed_factor,
        "turn_proxy": turn_proxy,
        "turn_norm": turn_norm,
        "ego_x1": ego_x1,
        "ego_x2": ego_x2,
        "ego_y1": ego_y1,
        "in_dynamic_ego_path": in_dynamic_ego_path,
        "on_turning_side": on_turning_side,
        "very_far": very_far,
        "far": far,
        "close": close,
        "very_close": very_close
    }

    return risk_score, risk_label, color, debug


# =============================
# PROCESS VIDEO
# =============================
cap = cv2.VideoCapture(video_path)
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

local_frame_idx = 0
shown_count = 0
saved_count = 0

print("Running dynamic ego-path risk assessment WITHOUT tracking...")

while local_frame_idx < num_frames_to_process:
    ret, frame = cap.read()

    if not ret:
        print("Could not read frame:", local_frame_idx)
        break

    original_frame = start_frame + local_frame_idx

    if local_frame_idx % frame_stride == 0:
        frame_h, frame_w = frame.shape[:2]
        vehicle_state = get_vehicle_state(original_frame)

        temp_path = "/content/temp_dynamic_risk.jpg"
        cv2.imwrite(temp_path, frame)

        # -----------------------------
        # YOLO detection, no tracking
        # -----------------------------
        yolo_results = yolo_model(
            temp_path,
            classes=[0],
            conf=conf_threshold,
            iou=iou_threshold,
            imgsz=1280,
            verbose=False
        )

        boxes = yolo_results[0].boxes.xyxy.cpu().numpy() if yolo_results[0].boxes is not None else []
        confs = yolo_results[0].boxes.conf.cpu().numpy() if yolo_results[0].boxes is not None else []

        output = frame.copy()

        # -----------------------------
        # Dynamic ego path drawing
        # -----------------------------
        turn_proxy = vehicle_state["heading_change"] + 0.2 * vehicle_state["gyroZ"]
        turn_norm = np.clip((turn_proxy / 4.0) * TURN_SIGN, -1.0, 1.0)

        base_center = frame_w * 0.50
        path_shift = turn_norm * frame_w * 0.18
        dynamic_center = base_center + path_shift

        base_width = frame_w * 0.40
        turn_extra_width = abs(turn_norm) * frame_w * 0.18
        path_width = base_width + turn_extra_width

        ego_x1 = int(max(0, dynamic_center - path_width / 2))
        ego_x2 = int(min(frame_w, dynamic_center + path_width / 2))
        ego_y1 = int(frame_h * 0.50)

        cv2.rectangle(output, (ego_x1, ego_y1), (ego_x2, frame_h), (255, 255, 255), 3)

        cv2.putText(
            output,
            "Dynamic ego path",
            (ego_x1, max(ego_y1 - 10, 25)),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (255, 255, 255),
            3
        )

        ego_text = (
            f"GPS: {vehicle_state['GPS_speed']:.1f} | "
            f"OBD: {vehicle_state['OBD_speed']:.1f} | "
            f"turn proxy: {turn_proxy:.2f}"
        )

        cv2.putText(
            output,
            ego_text,
            (20, 35),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (255, 255, 255),
            3
        )

        # -----------------------------
        # SAM2 segmentation
        # -----------------------------
        masks = []

        if len(boxes) > 0:
            sam_results = sam_model(
                temp_path,
                bboxes=np.array(boxes),
                verbose=False
            )

            if sam_results[0].masks is not None:
                masks = sam_results[0].masks.data.cpu().numpy()

        # -----------------------------
        # Risk per pedestrian
        # -----------------------------
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = box.astype(int)
            conf = float(confs[i]) if i < len(confs) else 0.0

            risk_score, risk_label, color, debug = calculate_risk_no_tracking_dynamic_path(
                x1, y1, x2, y2,
                frame_w, frame_h,
                vehicle_state
            )

            # SAM2 mask overlay
            if i < len(masks):
                mask_bool = masks[i].astype(bool)
                colored = output.copy()
                colored[mask_bool] = color
                output = cv2.addWeighted(output, 0.75, colored, 0.25, 0)

            # Bounding box
            cv2.rectangle(output, (x1, y1), (x2, y2), color, 3)

            # Centroid
            cx = int((x1 + x2) / 2)
            cy = int((y1 + y2) / 2)
            cv2.circle(output, (cx, cy), 5, color, -1)

            label = f"{risk_label} {risk_score:.2f} | conf {conf:.2f}"

            cv2.putText(
                output,
                label,
                (x1, max(y1 - 10, 25)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.65,
                (0, 0, 0),
                4
            )

            cv2.putText(
                output,
                label,
                (x1, max(y1 - 10, 25)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.65,
                color,
                2
            )

            debug_text = (
                f"P:{debug['proximity']:.2f} "
                f"E:{debug['ego_speed_factor']:.2f} "
                f"T:{debug['turn_norm']:.2f}"
            )

            cv2.putText(
                output,
                debug_text,
                (x1, min(y2 + 20, frame_h - 10)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.55,
                color,
                2
            )

        save_path = f"{output_folder}dynamic_risk_frame_{original_frame}.jpg"
        cv2.imwrite(save_path, output)
        saved_count += 1

        print(
            f"Frame {original_frame}: detections={len(boxes)}, "
            f"GPS={vehicle_state['GPS_speed']:.1f}, "
            f"OBD={vehicle_state['OBD_speed']:.1f}, "
            f"turn={turn_proxy:.2f}, "
            f"path_x=({ego_x1},{ego_x2})"
        )

        if shown_count < show_first_n:
            cv2_imshow(output)
            shown_count += 1

    local_frame_idx += 1

cap.release()

print("\nDone.")
print("Shown images:", shown_count)
print("Saved images:", saved_count)
print("Output folder:", output_folder)
```

```python
import cv2
import os
import numpy as np
import xml.etree.ElementTree as ET
import pandas as pd
from ultralytics import YOLO, SAM
from google.colab.patches import cv2_imshow

# =============================
# PATHS
# =============================
video_path = "/content/PIE/annotations/PIE_clips/set03/video_0006.mp4"
vehicle_xml_path = "/content/PIE/vehicle_annotations_unzipped/annotations_vehicle/set03/video_0006_obd.xml"

output_folder = "/content/PIE_video_0006_future_heading_dynamic_risk/"
os.makedirs(output_folder, exist_ok=True)

# =============================
# SETTINGS
# =============================
start_frame = 6870
num_frames_to_process = 622

conf_threshold = 0.50
iou_threshold = 0.50

frame_stride = 30
show_first_n = 25

# If the path shifts opposite to the actual future direction, change this to -1
TURN_SIGN = 1

# Future horizon for ego-path prediction
# 45 frames ≈ 1.5 seconds if video is 30 FPS
future_horizon = 45

print("Video exists:", os.path.exists(video_path))
print("Vehicle XML exists:", os.path.exists(vehicle_xml_path))

# =============================
# LOAD VEHICLE DATA
# =============================
tree = ET.parse(vehicle_xml_path)
root = tree.getroot()

vehicle_rows = []

for frame in root.findall("frame"):
    a = frame.attrib

    def safe_float(key, default=0.0):
        try:
            return float(a.get(key, default))
        except:
            return default

    frame_id = int(a["id"])

    vehicle_rows.append({
        "frame": frame_id,
        "GPS_speed": safe_float("GPS_speed"),
        "OBD_speed": safe_float("OBD_speed"),
        "gyroZ": safe_float("gyroZ"),
        "heading_angle": safe_float("heading_angle"),
        "yaw": safe_float("yaw"),
        "accX": safe_float("accX"),
        "accY": safe_float("accY"),
        "accZ": safe_float("accZ")
    })

df_vehicle = pd.DataFrame(vehicle_rows)

# =============================
# FUTURE HEADING / PATH DIRECTION
# =============================
def angle_difference(future_angle, current_angle):
    """
    Signed smallest angular difference in degrees.
    """
    return ((future_angle - current_angle + 180) % 360) - 180

# Frame-to-frame heading change
df_vehicle["heading_change"] = df_vehicle["heading_angle"].diff().fillna(0)
df_vehicle["heading_change"] = ((df_vehicle["heading_change"] + 180) % 360) - 180

# Future heading delta
future_deltas = []

for i in range(len(df_vehicle)):
    current_heading = df_vehicle.loc[i, "heading_angle"]
    future_i = min(i + future_horizon, len(df_vehicle) - 1)
    future_heading = df_vehicle.loc[future_i, "heading_angle"]

    delta = angle_difference(future_heading, current_heading)
    future_deltas.append(delta)

df_vehicle["future_heading_delta"] = future_deltas

# Smooth future heading delta so the ego path does not jump left/right
df_vehicle["future_heading_delta_smooth"] = (
    df_vehicle["future_heading_delta"]
    .rolling(window=31, center=True, min_periods=1)
    .mean()
)

# Normalize predicted future turn
df_vehicle["future_turn_norm"] = np.clip(
    df_vehicle["future_heading_delta_smooth"] / 25.0,
    -1.0,
    1.0
)

vehicle_data = {}

for _, row in df_vehicle.iterrows():
    vehicle_data[int(row["frame"])] = {
        "GPS_speed": float(row["GPS_speed"]),
        "OBD_speed": float(row["OBD_speed"]),
        "gyroZ": float(row["gyroZ"]),
        "heading_angle": float(row["heading_angle"]),
        "heading_change": float(row["heading_change"]),
        "future_heading_delta": float(row["future_heading_delta"]),
        "future_heading_delta_smooth": float(row["future_heading_delta_smooth"]),
        "future_turn_norm": float(row["future_turn_norm"]),
        "yaw": float(row["yaw"])
    }

def get_vehicle_state(frame_id):
    if frame_id in vehicle_data:
        return vehicle_data[frame_id]

    nearest_frame = min(vehicle_data.keys(), key=lambda x: abs(x - frame_id))
    return vehicle_data[nearest_frame]

print("Vehicle data loaded.")
print("Example frame 6870:", get_vehicle_state(6870))
print("Example frame 7170:", get_vehicle_state(7170))

# Optional: check future turn values for your segment
df_segment_vehicle = df_vehicle[
    (df_vehicle["frame"] >= start_frame) &
    (df_vehicle["frame"] <= start_frame + num_frames_to_process)
]

print("\nVehicle segment summary:")
display(df_segment_vehicle[[
    "frame",
    "GPS_speed",
    "OBD_speed",
    "heading_angle",
    "future_heading_delta_smooth",
    "future_turn_norm"
]].head())

# =============================
# LOAD MODELS
# =============================
yolo_model = YOLO("yolov8m.pt")
sam_model = SAM("sam2.1_s.pt")

# =============================
# RISK FUNCTION
# =============================
def calculate_zone_based_risk(
    x1, y1, x2, y2,
    frame_w, frame_h,
    vehicle_state,
    TURN_SIGN=1
):
    cx = (x1 + x2) / 2

    # Use pedestrian bottom point for zone decision
    foot_y = y2

    gps_speed = vehicle_state["GPS_speed"]
    obd_speed = vehicle_state["OBD_speed"]

    ego_speed = gps_speed if gps_speed > 0.1 else obd_speed
    ego_speed_factor = min(ego_speed / 50.0, 1.0)

    # Future heading direction from vehicle annotations
    turn_proxy = vehicle_state["future_heading_delta_smooth"]
    turn_norm = np.clip(vehicle_state["future_turn_norm"] * TURN_SIGN, -1.0, 1.0)

    # =============================
    # FUTURE-HEADING DYNAMIC EGO PATH
    # =============================
    # Slightly right-biased because this road scene is not perfectly centered
    base_center = frame_w * 0.56

    # Future heading shifts path toward expected direction of travel
    path_shift = turn_norm * frame_w * 0.32

    dynamic_center = base_center + path_shift

    # Wider corridor when future turn is stronger
    base_width = frame_w * 0.36
    turn_extra_width = abs(turn_norm) * frame_w * 0.22
    path_width = base_width + turn_extra_width

    ego_x1 = max(0, dynamic_center - path_width / 2)
    ego_x2 = min(frame_w, dynamic_center + path_width / 2)

    # Risk zones using foot position
    zone_top = frame_h * 0.48
    zone_mid = frame_h * 0.66
    zone_bottom = frame_h * 0.82

    in_ego_path = (ego_x1 <= cx <= ego_x2) and (foot_y >= zone_top)

    if turn_norm > 0.10:
        turning_side = 1
    elif turn_norm < -0.10:
        turning_side = -1
    else:
        turning_side = 0

    pedestrian_side = 1 if cx > frame_w / 2 else -1
    on_turning_side = turning_side != 0 and pedestrian_side == turning_side

    # =============================
    # ZONE-BASED RISK LOGIC
    # =============================
    if in_ego_path:
        if foot_y >= zone_bottom:
            risk_score = 0.90
            risk_label = "CRITICAL"
            color = (0, 0, 255)

        elif foot_y >= zone_mid:
            risk_score = 0.72
            risk_label = "HIGH"
            color = (0, 165, 255)

        else:
            risk_score = 0.50
            risk_label = "MEDIUM"
            color = (0, 255, 255)

    else:
        if on_turning_side and foot_y >= zone_mid:
            risk_score = 0.55
            risk_label = "MEDIUM"
            color = (0, 255, 255)

        elif on_turning_side and foot_y >= zone_top:
            risk_score = 0.42
            risk_label = "MEDIUM"
            color = (0, 255, 255)

        else:
            risk_score = 0.25
            risk_label = "LOW"
            color = (0, 255, 0)

    # Speed-based adjustment
    if ego_speed_factor > 0.30:
        if in_ego_path and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.90)
            risk_label = "CRITICAL"
            color = (0, 0, 255)

        elif in_ego_path and foot_y >= zone_mid:
            risk_score = max(risk_score, 0.78)
            risk_label = "HIGH"
            color = (0, 165, 255)

        elif on_turning_side and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.70)
            risk_label = "HIGH"
            color = (0, 165, 255)

    debug = {
        "ego_x1": ego_x1,
        "ego_x2": ego_x2,
        "zone_top": zone_top,
        "zone_mid": zone_mid,
        "zone_bottom": zone_bottom,
        "turn_proxy": turn_proxy,
        "turn_norm": turn_norm,
        "in_ego_path": in_ego_path,
        "on_turning_side": on_turning_side,
        "ego_speed": ego_speed,
        "ego_speed_factor": ego_speed_factor,
        "foot_y": foot_y
    }

    return risk_score, risk_label, color, debug

# =============================
# PROCESS VIDEO
# =============================
cap = cv2.VideoCapture(video_path)
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

local_frame_idx = 0
shown_count = 0
saved_count = 0

print("\nRunning future-heading dynamic risk assessment...")

while local_frame_idx < num_frames_to_process:
    ret, frame = cap.read()

    if not ret:
        print("Could not read frame:", local_frame_idx)
        break

    original_frame = start_frame + local_frame_idx

    if local_frame_idx % frame_stride == 0:
        frame_h, frame_w = frame.shape[:2]
        vehicle_state = get_vehicle_state(original_frame)

        temp_path = "/content/temp_future_heading_risk.jpg"
        cv2.imwrite(temp_path, frame)

        # YOLO detection
        yolo_results = yolo_model(
            temp_path,
            classes=[0],
            conf=conf_threshold,
            iou=iou_threshold,
            imgsz=1280,
            verbose=False
        )

        boxes = yolo_results[0].boxes.xyxy.cpu().numpy() if yolo_results[0].boxes is not None else []
        confs = yolo_results[0].boxes.conf.cpu().numpy() if yolo_results[0].boxes is not None else []

        output = frame.copy()

        # =============================
        # DRAW FUTURE-HEADING DYNAMIC EGO PATH
        # =============================
        turn_proxy = vehicle_state["future_heading_delta_smooth"]
        turn_norm = np.clip(vehicle_state["future_turn_norm"] * TURN_SIGN, -1.0, 1.0)

        base_center = frame_w * 0.56
        path_shift = turn_norm * frame_w * 0.32
        dynamic_center = base_center + path_shift

        base_width = frame_w * 0.36
        turn_extra_width = abs(turn_norm) * frame_w * 0.22
        path_width = base_width + turn_extra_width

        ego_x1 = int(max(0, dynamic_center - path_width / 2))
        ego_x2 = int(min(frame_w, dynamic_center + path_width / 2))

        zone_top = int(frame_h * 0.48)
        zone_mid = int(frame_h * 0.66)
        zone_bottom = int(frame_h * 0.82)

        cv2.rectangle(output, (ego_x1, zone_top), (ego_x2, frame_h), (255, 255, 255), 3)

        cv2.line(output, (ego_x1, zone_mid), (ego_x2, zone_mid), (0, 255, 255), 2)
        cv2.line(output, (ego_x1, zone_bottom), (ego_x2, zone_bottom), (0, 0, 255), 2)

        cv2.putText(
            output,
            "Predicted ego path",
            (ego_x1, max(zone_top - 10, 25)),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (255, 255, 255),
            3
        )

        cv2.putText(
            output,
            "MEDIUM",
            (ego_x1 + 10, zone_top + 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 255),
            2
        )

        cv2.putText(
            output,
            "HIGH",
            (ego_x1 + 10, zone_mid + 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 165, 255),
            2
        )

        cv2.putText(
            output,
            "CRITICAL",
            (ego_x1 + 10, zone_bottom + 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 0, 255),
            2
        )

        ego_text = (
            f"GPS: {vehicle_state['GPS_speed']:.1f} | "
            f"OBD: {vehicle_state['OBD_speed']:.1f} | "
            f"future heading: {turn_proxy:.2f} | "
            f"turn norm: {turn_norm:.2f}"
        )

        cv2.putText(
            output,
            ego_text,
            (20, 35),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.75,
            (255, 255, 255),
            3
        )

        # SAM2 segmentation
        masks = []

        if len(boxes) > 0:
            sam_results = sam_model(
                temp_path,
                bboxes=np.array(boxes),
                verbose=False
            )

            if sam_results[0].masks is not None:
                masks = sam_results[0].masks.data.cpu().numpy()

        # Risk assignment per pedestrian
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = box.astype(int)
            conf = float(confs[i]) if i < len(confs) else 0.0

            risk_score, risk_label, color, debug = calculate_zone_based_risk(
                x1, y1, x2, y2,
                frame_w, frame_h,
                vehicle_state,
                TURN_SIGN=TURN_SIGN
            )

            if i < len(masks):
                mask_bool = masks[i].astype(bool)
                colored = output.copy()
                colored[mask_bool] = color
                output = cv2.addWeighted(output, 0.75, colored, 0.25, 0)

            cv2.rectangle(output, (x1, y1), (x2, y2), color, 3)

            cx = int((x1 + x2) / 2)
            cy = int((y1 + y2) / 2)
            foot_x = cx
            foot_y = int(y2)

            cv2.circle(output, (cx, cy), 5, color, -1)
            cv2.circle(output, (foot_x, foot_y), 6, (255, 255, 255), -1)

            label = f"{risk_label} {risk_score:.2f} | conf {conf:.2f}"

            cv2.putText(
                output,
                label,
                (x1, max(y1 - 10, 25)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.65,
                (0, 0, 0),
                4
            )

            cv2.putText(
                output,
                label,
                (x1, max(y1 - 10, 25)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.65,
                color,
                2
            )

            small_text = f"speed:{debug['ego_speed']:.1f} turn:{debug['turn_norm']:.2f} foot:{debug['foot_y']:.0f}"

            cv2.putText(
                output,
                small_text,
                (x1, min(y2 + 20, frame_h - 10)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                color,
                2
            )

        save_path = f"{output_folder}future_heading_risk_frame_{original_frame}.jpg"
        cv2.imwrite(save_path, output)
        saved_count += 1

        print(
            f"Frame {original_frame}: detections={len(boxes)}, "
            f"GPS={vehicle_state['GPS_speed']:.1f}, "
            f"OBD={vehicle_state['OBD_speed']:.1f}, "
            f"future_heading={turn_proxy:.2f}, "
            f"turn_norm={turn_norm:.2f}, "
            f"path_x=({ego_x1},{ego_x2})"
        )

        if shown_count < show_first_n:
            cv2_imshow(output)
            shown_count += 1

    local_frame_idx += 1

cap.release()

print("\nDone.")
print("Shown images:", shown_count)
print("Saved images:", saved_count)
print("Output folder:", output_folder)
```

```python
import cv2
import os
import numpy as np
import xml.etree.ElementTree as ET
import pandas as pd
from ultralytics import YOLO, SAM
from google.colab.patches import cv2_imshow

# =============================
# PATHS
# =============================
video_path = "/content/PIE/annotations/PIE_clips/set03/video_0006.mp4"
vehicle_xml_path = "/content/PIE/vehicle_annotations_unzipped/annotations_vehicle/set03/video_0006_obd.xml"

output_folder = "/content/PIE_video_0006_annotation_driven_risk/"
os.makedirs(output_folder, exist_ok=True)

# =============================
# SETTINGS
# =============================
start_frame          = 6870
num_frames_to_process = 622
conf_threshold       = 0.50
iou_threshold        = 0.50
frame_stride         = 30
show_first_n         = 25
SCENE_WIDTH_METERS   = 10.0   # tune if corridor too wide/narrow
FUTURE_HORIZON_SEC   = 1.5    # how far ahead to project

print("Video exists:", os.path.exists(video_path))
print("Vehicle XML exists:", os.path.exists(vehicle_xml_path))

# =============================
# LOAD VEHICLE DATA
# =============================
tree = ET.parse(vehicle_xml_path)
root = tree.getroot()

vehicle_rows = []
for frame in root.findall("frame"):
    a = frame.attrib

    def safe_float(key, default=0.0):
        try:
            return float(a.get(key, default))
        except:
            return default

    vehicle_rows.append({
        "frame":         int(a["id"]),
        "GPS_speed":     safe_float("GPS_speed"),
        "OBD_speed":     safe_float("OBD_speed"),
        "gyroZ":         safe_float("gyroZ"),
        "heading_angle": safe_float("heading_angle"),
        "yaw":           safe_float("yaw"),
        "accX":          safe_float("accX"),
        "accY":          safe_float("accY"),
        "accZ":          safe_float("accZ")
    })

df = pd.DataFrame(vehicle_rows).sort_values("frame").reset_index(drop=True)

# =============================
# DERIVE EVERYTHING FROM ANNOTATIONS
# No hardcoded directions. No TURN_SIGN. No assumptions.
# The heading_angle sequence IS the ground truth path.
# =============================

def signed_heading_diff(h_future, h_current):
    """
    Smallest signed angular difference between two headings.
    Positive = turned RIGHT, Negative = turned LEFT.
    This handles the 359→1 wraparound correctly.
    """
    return ((h_future - h_current + 180) % 360) - 180

# How many frames ahead to look for future heading
# At 30 FPS, FUTURE_HORIZON_SEC=1.5 → 45 frames
FUTURE_FRAMES = int(30 * FUTURE_HORIZON_SEC)

future_heading_delta = []
for i in range(len(df)):
    current_h = df.loc[i, "heading_angle"]
    future_i  = min(i + FUTURE_FRAMES, len(df) - 1)
    future_h  = df.loc[future_i, "heading_angle"]
    delta = signed_heading_diff(future_h, current_h)
    future_heading_delta.append(delta)

df["future_heading_delta"] = future_heading_delta

# Smooth to remove GPS heading noise (your data has jumps like -11° then +9°
# within a few frames at low speed — those are GPS glitches, not real turns)
df["future_heading_delta_smooth"] = (
    df["future_heading_delta"]
    .rolling(window=21, center=True, min_periods=1)
    .mean()
)

# Also smooth gyroZ for the kinematic model
df["gyroZ_smooth"] = (
    df["gyroZ"]
    .rolling(window=15, center=True, min_periods=1)
    .mean()
)

# =============================
# BUILD LOOKUP
# =============================
vehicle_data = {}
for _, row in df.iterrows():
    vehicle_data[int(row["frame"])] = {
        "GPS_speed":                  float(row["GPS_speed"]),
        "OBD_speed":                  float(row["OBD_speed"]),
        "gyroZ":                      float(row["gyroZ"]),
        "gyroZ_smooth":               float(row["gyroZ_smooth"]),
        "heading_angle":              float(row["heading_angle"]),
        "future_heading_delta":       float(row["future_heading_delta"]),
        "future_heading_delta_smooth":float(row["future_heading_delta_smooth"]),
    }

def get_vehicle_state(frame_id):
    if frame_id in vehicle_data:
        return vehicle_data[frame_id]
    nearest = min(vehicle_data.keys(), key=lambda x: abs(x - frame_id))
    return vehicle_data[nearest]

print("Vehicle data loaded.")

# =============================
# VERIFY: print what the annotations actually say
# =============================
seg = df[(df["frame"] >= start_frame) & (df["frame"] <= start_frame + num_frames_to_process)]
print("\nWhat the annotations say about this segment:")
print(seg[["frame","GPS_speed","heading_angle",
           "future_heading_delta","future_heading_delta_smooth",
           "gyroZ_smooth"]].iloc[::30].to_string())

# =============================
# CORRIDOR: derived purely from annotation data
# =============================
def compute_ego_path(vehicle_state, frame_w, frame_h,
                     scene_width_m=10.0):
    """
    The corridor center is determined entirely by:
      1. future_heading_delta_smooth → WHERE the car will be going
         (read directly from heading_angle sequence in annotations)
      2. speed → HOW FAR it will travel (scales the lateral shift)
      3. gyroZ_smooth → instantaneous yaw rate (cross-check / blended)

    No hardcoded direction. No TURN_SIGN.
    If annotations say right turn → corridor goes right.
    If annotations say left turn  → corridor goes left.
    If annotations say straight   → corridor stays centered.
    """

    # Best speed in m/s
    speed_ms = vehicle_state["GPS_speed"] / 3.6
    if speed_ms < 0.5:
        speed_ms = vehicle_state["OBD_speed"] / 3.6

    is_stopped = speed_ms < (2.0 / 3.6)  # below 2 km/h = stopped

    # ── PRIMARY: future heading delta from annotations ──────────────────
    # This is the actual ground truth: how much will heading change
    # over the next FUTURE_HORIZON_SEC seconds, read from the XML.
    # Positive = right, Negative = left. Already smoothed.
    future_delta_deg = vehicle_state["future_heading_delta_smooth"]

    # Convert heading change to lateral displacement in meters.
    # If the car travels at speed_ms for FUTURE_HORIZON_SEC seconds
    # and turns by future_delta_deg, the arc lateral offset is:
    #   lateral = distance * sin(delta_rad)
    #   distance = speed_ms * time
    # For small angles sin(x)≈x but we use the full formula.
    if is_stopped:
        # Stopped: gyroZ is noise, heading GPS is noise → freeze corridor
        lateral_m = 0.0
    else:
        distance_m  = speed_ms * FUTURE_HORIZON_SEC
        delta_rad   = np.deg2rad(future_delta_deg)
        lateral_m   = distance_m * np.sin(delta_rad)
        # lateral_m > 0 = rightward, < 0 = leftward

    # ── MAP TO IMAGE ─────────────────────────────────────────────────────
    meters_to_px = frame_w / scene_width_m
    lateral_px   = lateral_m * meters_to_px

    # Base center: ego vehicle is slightly right of image center in this scene.
    # This is a camera mounting offset, not a direction assumption —
    # it is the same regardless of which way the car turns.
    base_center    = frame_w * 0.56
    dynamic_center = base_center + lateral_px   # purely from annotation data

    corridor_w = frame_w * 0.30
    ego_x1 = int(max(0, dynamic_center - corridor_w / 2))
    ego_x2 = int(min(frame_w, dynamic_center + corridor_w / 2))

    return {
        "ego_x1":        ego_x1,
        "ego_x2":        ego_x2,
        "center":        dynamic_center,
        "lateral_m":     lateral_m,
        "speed_ms":      speed_ms,
        "future_delta":  future_delta_deg,
        "is_stopped":    is_stopped,
        # Direction label read from annotations — not hardcoded
        "direction":     "RIGHT" if future_delta_deg > 2.0
                    else "LEFT"  if future_delta_deg < -2.0
                    else "STRAIGHT"
    }

# =============================
# RISK FUNCTION
# =============================
def calculate_risk(x1, y1, x2, y2, frame_w, frame_h, ego):
    cx     = (x1 + x2) / 2
    foot_y = y2

    speed_factor = min(ego["speed_ms"] / (50.0 / 3.6), 1.0)

    zone_top    = frame_h * 0.48
    zone_mid    = frame_h * 0.66
    zone_bottom = frame_h * 0.82

    in_corridor = (ego["ego_x1"] <= cx <= ego["ego_x2"]) and (foot_y >= zone_top)

    # Turning side derived from annotation direction — not hardcoded
    direction = ego["direction"]
    if direction == "RIGHT":
        turning_side = 1
    elif direction == "LEFT":
        turning_side = -1
    else:
        turning_side = 0

    pedestrian_side = 1 if cx > frame_w / 2 else -1
    on_turning_side = (turning_side != 0) and (pedestrian_side == turning_side)

    if in_corridor:
        if foot_y >= zone_bottom:
            risk_score = 0.90; risk_label = "CRITICAL"; color = (0, 0, 255)
        elif foot_y >= zone_mid:
            risk_score = 0.72; risk_label = "HIGH";     color = (0, 165, 255)
        else:
            risk_score = 0.50; risk_label = "MEDIUM";   color = (0, 255, 255)
    else:
        if on_turning_side and foot_y >= zone_mid:
            risk_score = 0.55; risk_label = "MEDIUM";   color = (0, 255, 255)
        elif on_turning_side and foot_y >= zone_top:
            risk_score = 0.42; risk_label = "MEDIUM";   color = (0, 255, 255)
        else:
            risk_score = 0.25; risk_label = "LOW";      color = (0, 255, 0)

    if speed_factor > 0.30:
        if in_corridor and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.90)
            risk_label = "CRITICAL"; color = (0, 0, 255)
        elif in_corridor and foot_y >= zone_mid:
            risk_score = max(risk_score, 0.78)
            risk_label = "HIGH";     color = (0, 165, 255)
        elif on_turning_side and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.70)
            risk_label = "HIGH";     color = (0, 165, 255)

    return risk_score, risk_label, color, {
        "in_corridor":     in_corridor,
        "on_turning_side": on_turning_side,
        "foot_y":          foot_y,
        "zone_top":        zone_top,
        "zone_mid":        zone_mid,
        "zone_bottom":     zone_bottom
    }

# =============================
# LOAD MODELS
# =============================
yolo_model = YOLO("yolov8m.pt")
sam_model  = SAM("sam2.1_s.pt")

# =============================
# PROCESS VIDEO
# =============================
cap = cv2.VideoCapture(video_path)
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

local_frame_idx = 0
shown_count     = 0
saved_count     = 0

print("\nRunning annotation-driven ego path risk assessment...")

while local_frame_idx < num_frames_to_process:
    ret, frame = cap.read()

    if not ret:
        print("Could not read frame:", local_frame_idx)
        break

    original_frame = start_frame + local_frame_idx

    if local_frame_idx % frame_stride == 0:
        frame_h, frame_w = frame.shape[:2]
        vehicle_state = get_vehicle_state(original_frame)

        # Corridor position comes entirely from annotation data
        ego = compute_ego_path(
            vehicle_state, frame_w, frame_h,
            scene_width_m=SCENE_WIDTH_METERS
        )

        ego_x1      = ego["ego_x1"]
        ego_x2      = ego["ego_x2"]
        zone_top    = int(frame_h * 0.48)
        zone_mid    = int(frame_h * 0.66)
        zone_bottom = int(frame_h * 0.82)

        temp_path = "/content/temp_risk.jpg"
        cv2.imwrite(temp_path, frame)

        # YOLO
        yolo_results = yolo_model(
            temp_path,
            classes=[0],
            conf=conf_threshold,
            iou=iou_threshold,
            imgsz=1280,
            verbose=False
        )

        boxes = yolo_results[0].boxes.xyxy.cpu().numpy() if yolo_results[0].boxes is not None else []
        confs = yolo_results[0].boxes.conf.cpu().numpy() if yolo_results[0].boxes is not None else []

        output = frame.copy()

        # Draw corridor
        cv2.rectangle(output,
                      (ego_x1, zone_top), (ego_x2, frame_h),
                      (255, 255, 255), 3)
        cv2.line(output, (ego_x1, zone_mid),    (ego_x2, zone_mid),    (0, 255, 255), 2)
        cv2.line(output, (ego_x1, zone_bottom), (ego_x2, zone_bottom), (0, 0, 255),   2)

        # Direction label — read from annotation, not assumed
        if ego["is_stopped"]:
            path_label = "STOPPED — corridor frozen"
        else:
            path_label = f"Ego path → {ego['direction']}"

        cv2.putText(output, path_label,
                    (ego_x1, max(zone_top - 10, 25)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 3)

        cv2.putText(output, "MEDIUM",
                    (ego_x1 + 10, zone_top + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        cv2.putText(output, "HIGH",
                    (ego_x1 + 10, zone_mid + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2)
        cv2.putText(output, "CRITICAL",
                    (ego_x1 + 10, zone_bottom + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

        # HUD — shows what the annotation data actually says
        hud = (
            f"heading:{vehicle_state['heading_angle']:.1f}°  "
            f"future_delta:{ego['future_delta']:.1f}°  "
            f"lateral:{ego['lateral_m']:.2f}m  "
            f"speed:{vehicle_state['GPS_speed']:.1f}km/h  "
            f"dir:{ego['direction']}"
        )
        cv2.putText(output, hud, (20, 35),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.62, (255, 255, 255), 3)

        # SAM2
        masks = []
        if len(boxes) > 0:
            sam_results = sam_model(temp_path, bboxes=np.array(boxes), verbose=False)
            if sam_results[0].masks is not None:
                masks = sam_results[0].masks.data.cpu().numpy()

        # Risk per pedestrian
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = box.astype(int)
            conf = float(confs[i]) if i < len(confs) else 0.0

            risk_score, risk_label, color, debug = calculate_risk(
                x1, y1, x2, y2,
                frame_w, frame_h, ego
            )

            if i < len(masks):
                mask_bool = masks[i].astype(bool)
                colored = output.copy()
                colored[mask_bool] = color
                output = cv2.addWeighted(output, 0.75, colored, 0.25, 0)

            cv2.rectangle(output, (x1, y1), (x2, y2), color, 3)
            cx = int((x1 + x2) / 2)
            cv2.circle(output, (cx, int((y1 + y2) / 2)), 5, color, -1)
            cv2.circle(output, (cx, int(y2)), 6, (255, 255, 255), -1)

            label = f"{risk_label} {risk_score:.2f} | conf {conf:.2f}"
            cv2.putText(output, label, (x1, max(y1 - 10, 25)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.65, (0, 0, 0), 4)
            cv2.putText(output, label, (x1, max(y1 - 10, 25)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.65, color, 2)

            small = (
                f"foot:{debug['foot_y']:.0f} "
                f"corridor:{'YES' if debug['in_corridor'] else 'NO'} "
                f"turn_side:{'YES' if debug['on_turning_side'] else 'NO'}"
            )
            cv2.putText(output, small,
                        (x1, min(y2 + 20, frame_h - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        save_path = f"{output_folder}risk_frame_{original_frame}.jpg"
        cv2.imwrite(save_path, output)
        saved_count += 1

        print(
            f"Frame {original_frame} | "
            f"heading:{vehicle_state['heading_angle']:.1f}° | "
            f"future_delta:{ego['future_delta']:.1f}° | "
            f"direction:{ego['direction']} | "
            f"lateral:{ego['lateral_m']:.2f}m | "
            f"corridor:({ego_x1},{ego_x2}) | "
            f"speed:{vehicle_state['GPS_speed']:.1f}km/h | "
            f"dets:{len(boxes)}"
        )

        if shown_count < show_first_n:
            cv2_imshow(output)
            shown_count += 1

    local_frame_idx += 1

cap.release()

print("\nDone.")
print("Shown:", shown_count)
print("Saved:", saved_count)
print("Output:", output_folder)
```

```python
import cv2
import os
import numpy as np
import xml.etree.ElementTree as ET
import pandas as pd
from ultralytics import YOLO, SAM
from google.colab.patches import cv2_imshow

# =============================
# PATHS
# =============================
video_path = "/content/PIE/annotations/PIE_clips/set03/video_0006.mp4"
vehicle_xml_path = "/content/PIE/vehicle_annotations_unzipped/annotations_vehicle/set03/video_0006_obd.xml"

output_folder = "/content/PIE_video_0006_annotation_driven_risk/"
os.makedirs(output_folder, exist_ok=True)

# =============================
# SETTINGS
# =============================
start_frame           = 6870
num_frames_to_process = 622
conf_threshold        = 0.50
iou_threshold         = 0.50
frame_stride          = 30
show_first_n          = 25
SCENE_WIDTH_METERS    = 10.0
FUTURE_HORIZON_SEC    = 1.5

print("Video exists:", os.path.exists(video_path))
print("Vehicle XML exists:", os.path.exists(vehicle_xml_path))

# =============================
# LOAD VEHICLE DATA
# =============================
tree = ET.parse(vehicle_xml_path)
root = tree.getroot()

vehicle_rows = []
for frame in root.findall("frame"):
    a = frame.attrib

    def safe_float(key, default=0.0):
        try:
            return float(a.get(key, default))
        except:
            return default

    vehicle_rows.append({
        "frame":         int(a["id"]),
        "GPS_speed":     safe_float("GPS_speed"),
        "OBD_speed":     safe_float("OBD_speed"),
        "gyroZ":         safe_float("gyroZ"),
        "heading_angle": safe_float("heading_angle"),
        "yaw":           safe_float("yaw"),
        "accX":          safe_float("accX"),
        "accY":          safe_float("accY"),
        "accZ":          safe_float("accZ")
    })

df = pd.DataFrame(vehicle_rows).sort_values("frame").reset_index(drop=True)

# =============================
# DERIVE DIRECTION FROM ANNOTATIONS
# =============================
def signed_heading_diff(h_future, h_current):
    """
    Signed angular difference. Handles 359->1 wraparound.
    Positive = RIGHT, Negative = LEFT.
    """
    return ((h_future - h_current + 180) % 360) - 180

FUTURE_FRAMES = int(30 * FUTURE_HORIZON_SEC)  # 45 frames at 30 FPS

future_heading_delta = []
for i in range(len(df)):
    current_h = df.loc[i, "heading_angle"]
    future_i  = min(i + FUTURE_FRAMES, len(df) - 1)
    future_h  = df.loc[future_i, "heading_angle"]
    delta = signed_heading_diff(future_h, current_h)
    future_heading_delta.append(delta)

df["future_heading_delta"] = future_heading_delta

# Smooth to remove GPS heading noise
df["future_heading_delta_smooth"] = (
    df["future_heading_delta"]
    .rolling(window=21, center=True, min_periods=1)
    .mean()
)

df["gyroZ_smooth"] = (
    df["gyroZ"]
    .rolling(window=15, center=True, min_periods=1)
    .mean()
)

# =============================
# BUILD LOOKUP
# =============================
vehicle_data = {}
for _, row in df.iterrows():
    vehicle_data[int(row["frame"])] = {
        "GPS_speed":                   float(row["GPS_speed"]),
        "OBD_speed":                   float(row["OBD_speed"]),
        "gyroZ":                       float(row["gyroZ"]),
        "gyroZ_smooth":                float(row["gyroZ_smooth"]),
        "heading_angle":               float(row["heading_angle"]),
        "future_heading_delta":        float(row["future_heading_delta"]),
        "future_heading_delta_smooth": float(row["future_heading_delta_smooth"]),
    }

def get_vehicle_state(frame_id):
    if frame_id in vehicle_data:
        return vehicle_data[frame_id]
    nearest = min(vehicle_data.keys(), key=lambda x: abs(x - frame_id))
    return vehicle_data[nearest]

print("Vehicle data loaded.")

seg = df[
    (df["frame"] >= start_frame) &
    (df["frame"] <= start_frame + num_frames_to_process)
]
print("\nAnnotation-derived directions for this segment:")
print(seg[[
    "frame", "GPS_speed", "heading_angle",
    "future_heading_delta", "future_heading_delta_smooth"
]].iloc[::30].to_string())

# =============================
# COMPUTE EGO PATH FROM ANNOTATIONS
# =============================
def compute_ego_path(vehicle_state, frame_w, frame_h,
                     scene_width_m=10.0):
    """
    Corridor position derived entirely from annotation data.
    future_heading_delta_smooth tells us where the car is heading.
    Positive delta = RIGHT turn -> corridor shifts RIGHT.
    Negative delta = LEFT turn  -> corridor shifts LEFT.
    """
    speed_ms = vehicle_state["GPS_speed"] / 3.6
    if speed_ms < 0.5:
        speed_ms = vehicle_state["OBD_speed"] / 3.6

    is_stopped = speed_ms < (2.0 / 3.6)

    future_delta_deg = vehicle_state["future_heading_delta_smooth"]

    if is_stopped:
        lateral_m = 0.0
    else:
        distance_m = speed_ms * FUTURE_HORIZON_SEC
        delta_rad  = np.deg2rad(future_delta_deg)
        lateral_m  = distance_m * np.sin(delta_rad)

    meters_to_px   = frame_w / scene_width_m
    lateral_px     = lateral_m * meters_to_px

    base_center    = frame_w * 0.56
    dynamic_center = base_center + lateral_px

    corridor_w = frame_w * 0.42

    ego_x1 = int(max(0, dynamic_center - corridor_w / 2))
    ego_x2 = int(min(frame_w, dynamic_center + corridor_w / 2))

    return {
        "ego_x1":       ego_x1,
        "ego_x2":       ego_x2,
        "center":       dynamic_center,
        "lateral_m":    lateral_m,
        "speed_ms":     speed_ms,
        "future_delta": future_delta_deg,
        "is_stopped":   is_stopped,
        "direction":    "RIGHT"    if future_delta_deg >  2.0
                   else "LEFT"     if future_delta_deg < -2.0
                   else "STRAIGHT"
    }

# =============================
# RISK FUNCTION
# =============================
def calculate_risk(x1, y1, x2, y2, frame_w, frame_h, ego):
    cx     = (x1 + x2) / 2
    foot_y = y2

    speed_factor = min(ego["speed_ms"] / (50.0 / 3.6), 1.0)

    # ---- ZONE THRESHOLDS ----
    zone_top    = frame_h * 0.55   # MEDIUM starts here (lowered)
    zone_mid    = frame_h * 0.72   # HIGH starts here
    zone_bottom = frame_h * 0.82   # CRITICAL starts here

    # ---- SIZE / DISTANCE PROXY ----
    box_height = y2 - y1
    box_area   = box_height * (x2 - x1)
    frame_area = frame_w * frame_h
    size_ratio = box_area / frame_area

    is_far      = size_ratio < 0.008   # very small box -> far away -> cap at LOW
    is_medium_d = size_ratio < 0.020   # medium distance -> cap at MEDIUM

    in_corridor = (ego["ego_x1"] <= cx <= ego["ego_x2"]) and (foot_y >= zone_top)

    direction = ego["direction"]
    if direction == "RIGHT":
        turning_side = 1
    elif direction == "LEFT":
        turning_side = -1
    else:
        turning_side = 0

    pedestrian_side = 1 if cx > frame_w / 2 else -1
    on_turning_side = (turning_side != 0) and (pedestrian_side == turning_side)

    if in_corridor:
        if foot_y >= zone_bottom:
            risk_score = 0.90; risk_label = "CRITICAL"; color = (0, 0, 255)
        elif foot_y >= zone_mid:
            risk_score = 0.72; risk_label = "HIGH";     color = (0, 165, 255)
        else:
            risk_score = 0.50; risk_label = "MEDIUM";   color = (0, 255, 255)
    else:
        if on_turning_side and foot_y >= zone_mid:
            risk_score = 0.55; risk_label = "MEDIUM";   color = (0, 255, 255)
        elif on_turning_side and foot_y >= zone_top:
            risk_score = 0.42; risk_label = "MEDIUM";   color = (0, 255, 255)
        else:
            risk_score = 0.25; risk_label = "LOW";      color = (0, 255, 0)

    if speed_factor > 0.30:
        if in_corridor and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.90)
            risk_label = "CRITICAL"; color = (0, 0, 255)
        elif in_corridor and foot_y >= zone_mid:
            risk_score = max(risk_score, 0.78)
            risk_label = "HIGH";     color = (0, 165, 255)
        elif on_turning_side and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.70)
            risk_label = "HIGH";     color = (0, 165, 255)

    # ---- DISTANCE OVERRIDE (size-based) ----
    if is_far:
        risk_score = 0.25; risk_label = "LOW"; color = (0, 255, 0)
    elif is_medium_d:
        if risk_label in ("CRITICAL", "HIGH"):
            risk_score = 0.50; risk_label = "MEDIUM"; color = (0, 255, 255)

    return risk_score, risk_label, color, {
        "in_corridor":     in_corridor,
        "on_turning_side": on_turning_side,
        "foot_y":          foot_y,
        "zone_top":        zone_top,
        "zone_mid":        zone_mid,
        "zone_bottom":     zone_bottom,
        "size_ratio":      size_ratio,
        "is_far":          is_far,
        "is_medium_d":     is_medium_d
    }

# =============================
# LOAD MODELS
# =============================
yolo_model = YOLO("yolov8m.pt")
sam_model  = SAM("sam2.1_s.pt")

# =============================
# PROCESS VIDEO
# =============================
cap = cv2.VideoCapture(video_path)
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

local_frame_idx = 0
shown_count     = 0
saved_count     = 0

print("\nRunning annotation-driven ego path risk assessment...")

while local_frame_idx < num_frames_to_process:
    ret, frame = cap.read()

    if not ret:
        print("Could not read frame:", local_frame_idx)
        break

    original_frame = start_frame + local_frame_idx

    if local_frame_idx % frame_stride == 0:
        frame_h, frame_w = frame.shape[:2]
        vehicle_state = get_vehicle_state(original_frame)

        ego = compute_ego_path(
            vehicle_state, frame_w, frame_h,
            scene_width_m=SCENE_WIDTH_METERS
        )

        ego_x1      = ego["ego_x1"]
        ego_x2      = ego["ego_x2"]

        # ---- ZONE THRESHOLDS ----
        zone_top    = int(frame_h * 0.55)   # MEDIUM starts here (lowered)
        zone_mid    = int(frame_h * 0.72)   # HIGH starts here
        zone_bottom = int(frame_h * 0.82)   # CRITICAL starts here

        temp_path = "/content/temp_risk.jpg"
        cv2.imwrite(temp_path, frame)

        # YOLO
        yolo_results = yolo_model(
            temp_path,
            classes=[0],
            conf=conf_threshold,
            iou=iou_threshold,
            imgsz=1280,
            verbose=False
        )

        boxes = yolo_results[0].boxes.xyxy.cpu().numpy() if yolo_results[0].boxes is not None else []
        confs = yolo_results[0].boxes.conf.cpu().numpy() if yolo_results[0].boxes is not None else []

        output = frame.copy()

        # --- Draw ego corridor ---
        cv2.rectangle(output,
                      (ego_x1, zone_top), (ego_x2, frame_h),
                      (255, 255, 255), 3)
        cv2.line(output,
                 (ego_x1, zone_mid), (ego_x2, zone_mid),
                 (0, 255, 255), 2)
        cv2.line(output,
                 (ego_x1, zone_bottom), (ego_x2, zone_bottom),
                 (0, 0, 255), 2)

        if ego["is_stopped"]:
            path_label = "STOPPED — corridor frozen"
        else:
            path_label = f"Ego path -> {ego['direction']}"

        cv2.putText(output, path_label,
                    (ego_x1, max(zone_top - 10, 25)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 3)

        cv2.putText(output, "MEDIUM",
                    (ego_x1 + 10, zone_top + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        cv2.putText(output, "HIGH",
                    (ego_x1 + 10, zone_mid + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2)
        cv2.putText(output, "CRITICAL",
                    (ego_x1 + 10, zone_bottom + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

        # HUD
        hud = (
            f"heading:{vehicle_state['heading_angle']:.1f}°  "
            f"future_delta:{ego['future_delta']:.1f}°  "
            f"lateral:{ego['lateral_m']:.2f}m  "
            f"speed:{vehicle_state['GPS_speed']:.1f}km/h  "
            f"dir:{ego['direction']}"
        )
        cv2.putText(output, hud, (20, 35),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.62, (255, 255, 255), 3)

        # SAM2
        masks = []
        if len(boxes) > 0:
            sam_results = sam_model(temp_path, bboxes=np.array(boxes), verbose=False)
            if sam_results[0].masks is not None:
                masks = sam_results[0].masks.data.cpu().numpy()

        # Risk per pedestrian
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = box.astype(int)
            conf = float(confs[i]) if i < len(confs) else 0.0

            risk_score, risk_label, color, debug = calculate_risk(
                x1, y1, x2, y2,
                frame_w, frame_h, ego
            )

            if i < len(masks):
                mask_bool = masks[i].astype(bool)
                colored = output.copy()
                colored[mask_bool] = color
                output = cv2.addWeighted(output, 0.75, colored, 0.25, 0)

            cv2.rectangle(output, (x1, y1), (x2, y2), color, 3)
            cx = int((x1 + x2) / 2)
            cv2.circle(output, (cx, int((y1 + y2) / 2)), 5, color, -1)
            cv2.circle(output, (cx, int(y2)), 6, (255, 255, 255), -1)

            label = f"{risk_label} {risk_score:.2f} | conf {conf:.2f}"
            cv2.putText(output, label,
                        (x1, max(y1 - 10, 25)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.65, (0, 0, 0), 4)
            cv2.putText(output, label,
                        (x1, max(y1 - 10, 25)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.65, color, 2)

            # Debug info including size_ratio for tuning
            small = (
                f"foot:{debug['foot_y']:.0f} "
                f"corridor:{'YES' if debug['in_corridor'] else 'NO'} "
                f"turn_side:{'YES' if debug['on_turning_side'] else 'NO'} "
                f"sz:{debug['size_ratio']:.4f}"
            )
            cv2.putText(output, small,
                        (x1, min(y2 + 20, frame_h - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        save_path = f"{output_folder}risk_frame_{original_frame}.jpg"
        cv2.imwrite(save_path, output)
        saved_count += 1

        print(
            f"Frame {original_frame} | "
            f"heading:{vehicle_state['heading_angle']:.1f}° | "
            f"future_delta:{ego['future_delta']:.1f}° | "
            f"direction:{ego['direction']} | "
            f"lateral:{ego['lateral_m']:.2f}m | "
            f"corridor:({ego_x1},{ego_x2}) | "
            f"speed:{vehicle_state['GPS_speed']:.1f}km/h | "
            f"dets:{len(boxes)}"
        )

        if shown_count < show_first_n:
            cv2_imshow(output)
            shown_count += 1

    local_frame_idx += 1

cap.release()

print("\nDone.")
print("Shown:", shown_count)
print("Saved:", saved_count)
print("Output:", output_folder)
```

```python
import cv2
import os
import numpy as np
import xml.etree.ElementTree as ET
import pandas as pd
from ultralytics import YOLO, SAM
from google.colab.patches import cv2_imshow

# =============================
# PATHS
# =============================
video_path = "/content/PIE/annotations/PIE_clips/set03/video_0006.mp4"
vehicle_xml_path = "/content/PIE/vehicle_annotations_unzipped/annotations_vehicle/set03/video_0006_obd.xml"

output_folder = "/content/PIE_video_0006_annotation_driven_risk/"
os.makedirs(output_folder, exist_ok=True)

# =============================
# SETTINGS
# =============================
start_frame           = 6870
num_frames_to_process = 622
conf_threshold        = 0.50
iou_threshold         = 0.50
frame_stride          = 30
show_first_n          = 25
SCENE_WIDTH_METERS    = 10.0
FUTURE_HORIZON_SEC    = 1.5

print("Video exists:", os.path.exists(video_path))
print("Vehicle XML exists:", os.path.exists(vehicle_xml_path))

# =============================
# LOAD VEHICLE DATA
# =============================
tree = ET.parse(vehicle_xml_path)
root = tree.getroot()

vehicle_rows = []
for frame in root.findall("frame"):
    a = frame.attrib

    def safe_float(key, default=0.0):
        try:
            return float(a.get(key, default))
        except:
            return default

    vehicle_rows.append({
        "frame":         int(a["id"]),
        "GPS_speed":     safe_float("GPS_speed"),
        "OBD_speed":     safe_float("OBD_speed"),
        "gyroZ":         safe_float("gyroZ"),
        "heading_angle": safe_float("heading_angle"),
        "yaw":           safe_float("yaw"),
        "accX":          safe_float("accX"),
        "accY":          safe_float("accY"),
        "accZ":          safe_float("accZ")
    })

df = pd.DataFrame(vehicle_rows).sort_values("frame").reset_index(drop=True)

# =============================
# DERIVE DIRECTION FROM ANNOTATIONS
# =============================
def signed_heading_diff(h_future, h_current):
    return ((h_future - h_current + 180) % 360) - 180

FUTURE_FRAMES = int(30 * FUTURE_HORIZON_SEC)

future_heading_delta = []
for i in range(len(df)):
    current_h = df.loc[i, "heading_angle"]
    future_i  = min(i + FUTURE_FRAMES, len(df) - 1)
    future_h  = df.loc[future_i, "heading_angle"]
    delta = signed_heading_diff(future_h, current_h)
    future_heading_delta.append(delta)

df["future_heading_delta"] = future_heading_delta

df["future_heading_delta_smooth"] = (
    df["future_heading_delta"]
    .rolling(window=21, center=True, min_periods=1)
    .mean()
)

df["gyroZ_smooth"] = (
    df["gyroZ"]
    .rolling(window=15, center=True, min_periods=1)
    .mean()
)

# =============================
# BUILD LOOKUP
# =============================
vehicle_data = {}
for _, row in df.iterrows():
    vehicle_data[int(row["frame"])] = {
        "GPS_speed":                   float(row["GPS_speed"]),
        "OBD_speed":                   float(row["OBD_speed"]),
        "gyroZ":                       float(row["gyroZ"]),
        "gyroZ_smooth":                float(row["gyroZ_smooth"]),
        "heading_angle":               float(row["heading_angle"]),
        "future_heading_delta":        float(row["future_heading_delta"]),
        "future_heading_delta_smooth": float(row["future_heading_delta_smooth"]),
    }

def get_vehicle_state(frame_id):
    if frame_id in vehicle_data:
        return vehicle_data[frame_id]
    nearest = min(vehicle_data.keys(), key=lambda x: abs(x - frame_id))
    return vehicle_data[nearest]

print("Vehicle data loaded.")

seg = df[
    (df["frame"] >= start_frame) &
    (df["frame"] <= start_frame + num_frames_to_process)
]
print("\nAnnotation-derived directions for this segment:")
print(seg[[
    "frame", "GPS_speed", "heading_angle",
    "future_heading_delta", "future_heading_delta_smooth"
]].iloc[::30].to_string())

# =============================
# COMPUTE EGO PATH FROM ANNOTATIONS
# =============================
def compute_ego_path(vehicle_state, frame_w, frame_h, scene_width_m=10.0):
    speed_ms = vehicle_state["GPS_speed"] / 3.6
    if speed_ms < 0.5:
        speed_ms = vehicle_state["OBD_speed"] / 3.6

    is_stopped = speed_ms < (2.0 / 3.6)
    future_delta_deg = vehicle_state["future_heading_delta_smooth"]

    if is_stopped:
        lateral_m = 0.0
    else:
        distance_m = speed_ms * FUTURE_HORIZON_SEC
        delta_rad  = np.deg2rad(future_delta_deg)
        lateral_m  = distance_m * np.sin(delta_rad)

    meters_to_px   = frame_w / scene_width_m
    lateral_px     = lateral_m * meters_to_px
    base_center    = frame_w * 0.56
    dynamic_center = base_center + lateral_px
    corridor_w     = frame_w * 0.42

    ego_x1 = int(max(0, dynamic_center - corridor_w / 2))
    ego_x2 = int(min(frame_w, dynamic_center + corridor_w / 2))

    return {
        "ego_x1":       ego_x1,
        "ego_x2":       ego_x2,
        "center":       dynamic_center,
        "lateral_m":    lateral_m,
        "speed_ms":     speed_ms,
        "future_delta": future_delta_deg,
        "is_stopped":   is_stopped,
        "direction":    "RIGHT"    if future_delta_deg >  2.0
                   else "LEFT"     if future_delta_deg < -2.0
                   else "STRAIGHT"
    }

# =============================
# RISK FUNCTION
# =============================
def calculate_risk(x1, y1, x2, y2, frame_w, frame_h, ego):
    cx     = (x1 + x2) / 2
    foot_y = y2

    speed_factor = min(ego["speed_ms"] / (50.0 / 3.6), 1.0)

    # ---- ZONE THRESHOLDS ----
    zone_top    = frame_h * 0.55   # MEDIUM starts here
    zone_mid    = frame_h * 0.72   # HIGH starts here
    zone_bottom = frame_h * 0.82   # CRITICAL starts here

    # ---- SIZE / DISTANCE PROXY ----
    box_height = y2 - y1
    box_area   = box_height * (x2 - x1)
    frame_area = frame_w * frame_h
    size_ratio = box_area / frame_area

    # Raised thresholds so close people are never wrongly downgraded
    is_far      = size_ratio < 0.005   # only truly tiny far-away boxes
    is_medium_d = size_ratio < 0.012   # only genuinely medium-distance boxes

    in_corridor = (ego["ego_x1"] <= cx <= ego["ego_x2"]) and (foot_y >= zone_top)

    direction = ego["direction"]
    if direction == "RIGHT":
        turning_side = 1
    elif direction == "LEFT":
        turning_side = -1
    else:
        turning_side = 0

    pedestrian_side = 1 if cx > frame_w / 2 else -1
    on_turning_side = (turning_side != 0) and (pedestrian_side == turning_side)

    if in_corridor:
        if foot_y >= zone_bottom:
            risk_score = 0.90; risk_label = "CRITICAL"; color = (0, 0, 255)
        elif foot_y >= zone_mid:
            risk_score = 0.72; risk_label = "HIGH";     color = (0, 165, 255)
        else:
            risk_score = 0.50; risk_label = "MEDIUM";   color = (0, 255, 255)
    else:
        if on_turning_side and foot_y >= zone_mid:
            risk_score = 0.55; risk_label = "MEDIUM";   color = (0, 255, 255)
        elif on_turning_side and foot_y >= zone_top:
            risk_score = 0.42; risk_label = "MEDIUM";   color = (0, 255, 255)
        else:
            risk_score = 0.25; risk_label = "LOW";      color = (0, 255, 0)

    if speed_factor > 0.30:
        if in_corridor and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.90)
            risk_label = "CRITICAL"; color = (0, 0, 255)
        elif in_corridor and foot_y >= zone_mid:
            risk_score = max(risk_score, 0.78)
            risk_label = "HIGH";     color = (0, 165, 255)
        elif on_turning_side and foot_y >= zone_bottom:
            risk_score = max(risk_score, 0.70)
            risk_label = "HIGH";     color = (0, 165, 255)

    # ---- DISTANCE OVERRIDE (size-based) ----
    # Only downgrade truly small/far boxes — close people are never touched
    if is_far:
        risk_score = 0.25; risk_label = "LOW"; color = (0, 255, 0)
    elif is_medium_d:
        if risk_label in ("CRITICAL", "HIGH"):
            risk_score = 0.50; risk_label = "MEDIUM"; color = (0, 255, 255)

    return risk_score, risk_label, color, {
        "in_corridor":     in_corridor,
        "on_turning_side": on_turning_side,
        "foot_y":          foot_y,
        "zone_top":        zone_top,
        "zone_mid":        zone_mid,
        "zone_bottom":     zone_bottom,
        "size_ratio":      size_ratio,
        "is_far":          is_far,
        "is_medium_d":     is_medium_d
    }

# =============================
# LOAD MODELS
# =============================
yolo_model = YOLO("yolov8m.pt")
sam_model  = SAM("sam2.1_s.pt")

# =============================
# PROCESS VIDEO
# =============================
cap = cv2.VideoCapture(video_path)
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

local_frame_idx = 0
shown_count     = 0
saved_count     = 0

print("\nRunning annotation-driven ego path risk assessment...")

while local_frame_idx < num_frames_to_process:
    ret, frame = cap.read()

    if not ret:
        print("Could not read frame:", local_frame_idx)
        break

    original_frame = start_frame + local_frame_idx

    if local_frame_idx % frame_stride == 0:
        frame_h, frame_w = frame.shape[:2]
        vehicle_state = get_vehicle_state(original_frame)

        ego = compute_ego_path(
            vehicle_state, frame_w, frame_h,
            scene_width_m=SCENE_WIDTH_METERS
        )

        ego_x1      = ego["ego_x1"]
        ego_x2      = ego["ego_x2"]

        # ---- ZONE THRESHOLDS ----
        zone_top    = int(frame_h * 0.55)
        zone_mid    = int(frame_h * 0.72)
        zone_bottom = int(frame_h * 0.82)

        temp_path = "/content/temp_risk.jpg"
        cv2.imwrite(temp_path, frame)

        # YOLO
        yolo_results = yolo_model(
            temp_path,
            classes=[0],
            conf=conf_threshold,
            iou=iou_threshold,
            imgsz=1280,
            verbose=False
        )

        boxes = yolo_results[0].boxes.xyxy.cpu().numpy() if yolo_results[0].boxes is not None else []
        confs = yolo_results[0].boxes.conf.cpu().numpy() if yolo_results[0].boxes is not None else []

        output = frame.copy()

        # --- Draw ego corridor ---
        cv2.rectangle(output,
                      (ego_x1, zone_top), (ego_x2, frame_h),
                      (255, 255, 255), 3)
        cv2.line(output,
                 (ego_x1, zone_mid), (ego_x2, zone_mid),
                 (0, 255, 255), 2)
        cv2.line(output,
                 (ego_x1, zone_bottom), (ego_x2, zone_bottom),
                 (0, 0, 255), 2)

        if ego["is_stopped"]:
            path_label = "STOPPED — corridor frozen"
        else:
            path_label = f"Ego path -> {ego['direction']}"

        cv2.putText(output, path_label,
                    (ego_x1, max(zone_top - 10, 25)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 3)

        cv2.putText(output, "MEDIUM",
                    (ego_x1 + 10, zone_top + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        cv2.putText(output, "HIGH",
                    (ego_x1 + 10, zone_mid + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2)
        cv2.putText(output, "CRITICAL",
                    (ego_x1 + 10, zone_bottom + 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

        # HUD
        hud = (
            f"heading:{vehicle_state['heading_angle']:.1f}°  "
            f"future_delta:{ego['future_delta']:.1f}°  "
            f"lateral:{ego['lateral_m']:.2f}m  "
            f"speed:{vehicle_state['GPS_speed']:.1f}km/h  "
            f"dir:{ego['direction']}"
        )
        cv2.putText(output, hud, (20, 35),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.62, (255, 255, 255), 3)

        # SAM2
        masks = []
        if len(boxes) > 0:
            sam_results = sam_model(temp_path, bboxes=np.array(boxes), verbose=False)
            if sam_results[0].masks is not None:
                masks = sam_results[0].masks.data.cpu().numpy()

        # Risk per pedestrian
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = box.astype(int)
            conf = float(confs[i]) if i < len(confs) else 0.0

            risk_score, risk_label, color, debug = calculate_risk(
                x1, y1, x2, y2,
                frame_w, frame_h, ego
            )

            if i < len(masks):
                mask_bool = masks[i].astype(bool)
                colored = output.copy()
                colored[mask_bool] = color
                output = cv2.addWeighted(output, 0.75, colored, 0.25, 0)

            cv2.rectangle(output, (x1, y1), (x2, y2), color, 3)
            cx = int((x1 + x2) / 2)
            cv2.circle(output, (cx, int((y1 + y2) / 2)), 5, color, -1)
            cv2.circle(output, (cx, int(y2)), 6, (255, 255, 255), -1)

            label = f"{risk_label} {risk_score:.2f} | conf {conf:.2f}"
            cv2.putText(output, label,
                        (x1, max(y1 - 10, 25)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.65, (0, 0, 0), 4)
            cv2.putText(output, label,
                        (x1, max(y1 - 10, 25)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.65, color, 2)

            small = (
                f"foot:{debug['foot_y']:.0f} "
                f"corridor:{'YES' if debug['in_corridor'] else 'NO'} "
                f"turn_side:{'YES' if debug['on_turning_side'] else 'NO'} "
                f"sz:{debug['size_ratio']:.4f}"
            )
            cv2.putText(output, small,
                        (x1, min(y2 + 20, frame_h - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        save_path = f"{output_folder}risk_frame_{original_frame}.jpg"
        cv2.imwrite(save_path, output)
        saved_count += 1

        print(
            f"Frame {original_frame} | "
            f"heading:{vehicle_state['heading_angle']:.1f}° | "
            f"future_delta:{ego['future_delta']:.1f}° | "
            f"direction:{ego['direction']} | "
            f"lateral:{ego['lateral_m']:.2f}m | "
            f"corridor:({ego_x1},{ego_x2}) | "
            f"speed:{vehicle_state['GPS_speed']:.1f}km/h | "
            f"dets:{len(boxes)}"
        )

        if shown_count < show_first_n:
            cv2_imshow(output)
            shown_count += 1

    local_frame_idx += 1

cap.release()

print("\nDone.")
print("Shown:", shown_count)
print("Saved:", saved_count)
print("Output:", output_folder)
```
