```python
import xml.etree.ElementTree as ET
from collections import Counter

vehicle_xml_path = "/content/jaad_repo/annotations_vehicle/video_0135_vehicle.xml"

tree = ET.parse(vehicle_xml_path)
root = tree.getroot()

tag_counter = Counter()
attribute_counter = Counter()

for elem in root.iter():
    tag_counter[elem.tag] += 1
    for key in elem.attrib.keys():
        attribute_counter[key] += 1

print("Root tag:", root.tag)
print("Root attributes:", root.attrib)

print("\nTags:")
for tag, count in tag_counter.most_common():
    print(tag, count)

print("\nAttribute keys:")
for key, count in attribute_counter.most_common():
    print(key, count)
```
